```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
processus = pd.read_excel("Classeur1.xlsx")
```

```python
processus
```

```python
thread = pd.read_excel("dataset_thread.xlsx") 
```

```python
thread
```

```python
processus["Taux_d'echec_L1(en %)"] = processus["Taux_d'echec_L1(en %)"]*100
processus["Taux_d'echec_L2(en %)"] = round(processus["Taux_d'echec_L2(en %)"],2)
```

```python
thread["Taux_d'echec_L2(en %)"] = round(thread["Taux_d'echec_L2(en %)"],2)
thread["Taux_d'echec_L1(en %)"] = thread["Taux_d'echec_L1(en %)"]*100
```

```python
processus
```

```python
thread
```

```python
mean1_proc = np.mean(processus["Taux_d'echec_L1(en %)"])
mean2_proc = np.mean(processus["Taux_d'echec_L2(en %)"])
mean1_thre = np.mean(thread["Taux_d'echec_L1(en %)"])
mean2_thre = np.mean(thread["Taux_d'echec_L2(en %)"])
print("La moyenne du taux d'echec L1 des PROCESSUS est de ",mean1_proc)
print("La moyenne du taux d'echec L2 des PROCESSUS est de ",mean2_proc)
print("La moyenne du taux d'echec L1 des THREAD est de ",mean1_thre)
print("La moyenne du taux d'echec L2 des THREAD est de ",mean2_thre)
```

```python
processus[["Cache_Miss_L1","Cache_Misses_L2"]].boxplot(
    figsize = (4,4),
    widths = 0.3                                        )
figsize = (3,1)
plt.title("Boxplot des caches misses des processus")
plt.show()
```

```python
thread[["Cache_Miss_L1","Cache_Misses_L2"]].boxplot(
    figsize = (4,4),
    widths = 0.3                                       )
figsize = (3,1)
plt.title("Boxplot des caches misses des thread")
plt.show()
```

```python


# 2. Calcul de la matrice de corrélation
# Par défaut, la méthode utilise la corrélation de Pearson
matrice_corr = processus.corr()
print(matrice_corr)

# 3. Visualisation sous forme de Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(matrice_corr, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)

plt.title("Matrice de corrélation des processus")
plt.savefig("heatmap_processus", bbox_inches = 'tight')
plt.show()
```

```python
# 2. Calcul de la matrice de corrélation
# Par défaut, la méthode utilise la corrélation de Pearson
matrice_corr = thread.corr()
print(matrice_corr)

# 3. Visualisation sous forme de Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(matrice_corr, annot=True, cmap='seismic', fmt=".2f", vmin=-1, vmax=1)

plt.title("Matrice de corrélation des thread")
plt.savefig("heatmap_thread", bbox_inches = 'tight')
plt.show()
```

```python
plt.plot(
    processus["Cache_Miss_L1"],
     processus["Nbre_Changement_de_contexte"],
    label = "Cas processus ",
    alpha = 0.6,
    marker = "o",
    linestyle = "-",
    markersize = 6,
    linewidth = 2,
)
plt.plot(
    thread["Cache_Miss_L1"],
    thread["Nbre_Changement_de_contexte"],
    label = "Cas thread",
    alpha = 0.6,
    marker = "x",
    linestyle = "-",
    markersize = 6,
    linewidth = 2,
)

plt.xlabel("Cache misses L1")
plt.ylabel("Nbr changement de contexte")
plt.title("Comparaison cache misses L1 processus vs thread")
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig("scaterplot_L1", bbox_inches = 'tight')
plt.show()
```

```python
plt.plot(
    processus["Cache_Misses_L2"],
     processus["Nbre_Changement_de_contexte"],
    label = "Cas processus ",
    alpha = 0.6,
    marker = "o",
    linestyle = "-",
    markersize = 6,
    linewidth = 2,
)
plt.plot(
    thread["Cache_Misses_L2"],
    thread["Nbre_Changement_de_contexte"],
    label = "Cas thread",
    alpha = 0.6,
    marker = "x",
    linestyle = "-",
    markersize = 6,
    linewidth = 2,
)

plt.xlabel("Cache misses L2")
plt.ylabel("Nbr changement de contexte")
plt.title("Comparaison cache misses L2 processus vs thread")
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig("scaterplot_L2", bbox_inches = 'tight')
plt.show()
```
